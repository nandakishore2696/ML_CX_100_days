## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(tips, x='total_bill',y='tip', hue='sex', style='smoker',size='size')
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.columns
```

```python
sns.barplot(titanic,x='pclass',y='age')
```

```python
sns.barplot(titanic,x='pclass',y='age',hue='sex')
```

```python
sns.barplot(titanic,x='pclass',y='fare')
```

```python
sns.barplot(titanic,x='pclass',y='fare',hue='sex')
```

## 3. Box Plot (Numerical - Categorical)

```python
titanic.columns
```

```python
sns.boxplot(titanic, x='sex', y='age')
```

```python
sns.boxplot(titanic, x='sex', y='age', hue='survived')
```

## 4. Distplot (Numerical - Categorical)

```python
sns.displot(titanic, x='age', hue='sex',kind='kde')
```

```python
sns.displot(titanic, x='age', hue='survived',kind='kde')
```

## 5. HeatMap (Categorical - Categorical)

```python
pd.crosstab(titanic['pclass'],titanic['survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```

```python
titanic.groupby('pclass')['survived'].mean()*100
```

```python
(titanic.groupby('pclass')['survived'].mean()*100).plot(kind='bar')
```

```python
(titanic.groupby('sex')['survived'].mean()*100)
```

```python
(titanic.groupby('embarked')['survived'].mean()*100)
```

## 6. ClusterMap (Categorical - Categorical)

```python
titanic.columns
```

```python
pd.crosstab(titanic['sibsp'],titanic['survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['parch'],titanic['survived']))
```

## 7. Pairplot

```python
sns.pairplot(iris,hue='species')
```
